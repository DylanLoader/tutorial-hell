## Fundamentals of Graphs

A **Graph** is a collection of vertices and edges. Also denoted G(V,E). 

A **directed graph** has edges that have a defined direction. In an **Undirected Graph** nodes are connected by edges, but there is no direction to their edge. A weighted graph is a graph where each edge has a weight associated with it. 

A **complete graph** is a graph in which each node is connected with an edge to every other node in the graph. 

Graphs are difficult to understand as they cannot be represented in Eudlidean Space. 

Benefits of Graphs:
- Give an intuitive rep to abstract concepts
- Intuitively visually representation of information
- Form a natural basis for analyzing relationships in a social context
- Break down complex problems into simpler representations
- Transform the complex problems into representations from different perspectives

Traditional Graph Analysis Methods
- Searching algorithms BFS, DFS
- Shortest path algorithms eg Dijkstra's algorithm, Nearest Neighbour
- Spanning tree algorithms eg. Prim's algorithm
- Clustering methods eg. Highly Connected Components, k-means

These approaches have limited applications, which is why we need GNNs to generalize the power of graphs to other use cases.



## Mathematical Representation of Graphs

Graph: $G = (V, E)$

Vertices: $V = {A, B, C, D, E}$

Edges: 
$$
E = \{AB, BC, CD, DE\}\\
E = \{(A,B), (B,C), (C,D), (D,E)\}
$$

For weighted edges we add a weight to the representation such as, $E = {(A, B, 10), (C, D, 12)}$

Two nodes are **neighbours** if they are connected with an edge

#### Adjacency Matrix

An Adjacency matrix is a 2D square matrix
- Each node in the graph is the same along each axis. 
- Unweighted graphs have binary entries of 1 or 0
- Weighted graphs have numeric entries where missing weights are represented by -1
- In directed graphs the AM need not be symmetric

A **Complete Graph** is a graph where we have: 

- All elementst of adjacency matrix $\mathbb{A}$ are 1/T, except along the diagonal
- Path exist for each and every node

A **Sparse Graph** has:
- Not all elements of adjacency matrix $\mathbb{A}$ are 1/T, lots of values are -1 or False or 0. 
- Not every node is connected to each other

Node attribute matrix or Feature Matrix (X)
- The features or attributes of each node
- A graph with N nodes with the size of node attributes as F, 
  - Matrix shape = N x F
